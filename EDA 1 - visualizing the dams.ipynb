```python
def visualize(data): 
    positions = [6071, 6304, 7026, 7629, 7767, 8944, 11107] 
    plt.figure(figsize = (20,20))
    for i in range(5): 
        plt.subplot(1,5,i+1)
        plt.imshow(data[:,:,i])
        for pos in positions: 
            row = pos // 120 
            col = pos - 120*row 
            plt.scatter([row],[col], color = 'red', s = 40)

            
train_files = [os.path.join('./train',x) for x in os.listdir('./train')]
print("Visualizing the first 20 data points with positions of dams highlighted")
for i in range(20): 
    data = np.load(train_files[i]) 
    visualize(data)
```
